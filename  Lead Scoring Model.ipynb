## Importing Required Libraries

```python
#Importing Libraries and reading the data  
import numpy as np
import pandas as pd

data=pd.read_csv('Data_Science_Internship - Dump.csv')
df = pd.DataFrame(data)
df
```

## Data Cleaning ( including Feature Selection)

```python
#Filling the values of all empty space 
df["lost_reason"].fillna("No lost reason given", inplace = True)
df["budget"].fillna("0", inplace = True)
df["lease"].fillna("0", inplace = True)
df["room_type"].fillna("No room type given", inplace = True)
df["movein"].fillna("01/09/22", inplace = True)
df.info()
```

```python
#Data which leads with STATUS other than ‘WON’ or ‘LOST’ dropped.
indexAge = df[ (df['status'] == 'OPPORTUNITY') ].index
df.drop(indexAge , inplace=True)
df.info()
```

```python
# Treating all coulmns as CATEGORICAL column
df.astype('category')
```

```python
df.info()
```

```python
#Converting Dates as Datetime 
df["movein"] = pd.to_datetime(df["movein"])
df
df.dtypes
```

```python
#converting lease into integer datatype

def my_func(s):
    # perform some operation on x
    if(s=='semester-stay'):
        return 24
    elif(s=='full-course-stay' or 'full-year-stay'):
        return 44
    else:
        s=s.split(' ')
        l=[0]
        for i in s:
            if i.isdigit():
                l.append(int(i))
        return max(l)

df['lease'] = df['lease'].apply(my_func)
df.head(9)
```

```python
#converting budget as integer 

import re
def func(s):
    s= s+' 0'
    matches = re.findall(r'\d+',s)
    max_value = max(int(m) for m in matches)
    return max_value

df['budget'] = df['budget'].apply(func)
df
```

```python
#checking the type of data 
df['budget'] = pd.to_numeric(df['budget'], errors='coerce').astype(int)
df.dtypes
```
